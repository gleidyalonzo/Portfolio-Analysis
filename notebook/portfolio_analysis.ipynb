```python
#Importing necessary liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
```

# Portfolio Analysis

```python
#load data
low_data = pd.read_csv('../data/low_risk_port.csv')
low_data.head(10)
```

```python
#convertinf  date column to datetime datatype
low_data['date'] = pd.to_datetime(low_data['date'])
#setting date column to index
low_data.set_index('date', inplace=True)
```

#### Individual Asset Analysis

```python
# Calculate daily returns
returns = low_data.pivot(columns='name', values='close').pct_change()

# Individual Asset Analysis
mean_returns = returns.mean()
variance = returns.var()
standard_deviation = returns.std()

# Print Individual Asset Analysis
print("Individual Asset Analysis:")
for asset in returns.columns:
    print("Asset:", asset)
    print("Average Return:", mean_returns[asset])
    print("Variance:", variance[asset])
    print("Standard Deviation:", standard_deviation[asset])
    print()

# Portfolio Return
weights = np.array([1 / len(returns.columns) for _ in range(len(returns.columns))])# Equal weights for simplicity
portfolio_return = np.dot(mean_returns, weights)

# Portfolio Risk
cov_matrix = returns.cov()
portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Print Portfolio Return and Portfolio Risk
print("Overall Portfolio Return:", portfolio_return)
print(" Overall Portfolio Risk:", portfolio_risk)
```

#### Key Takeaway:

- Among the stocks, Microsoft has the highest return (0.00104), which indicates this stocks perform better over the period between 2013-2018. Procter & Gamble had the lowest average return.

- Microsoft has the highest volatility compare to other assets, where its standard deviation and variance is the high among all stocks. PepsiCo has the lowest variance and standard deviation, which make it relatively lower risk.

- Overall, the portfolio has a return on 0.00052 and a low risk with SD of 0.00689, which makes it a balanced portfolio with diversified risk.

- Visa and #M have a better risk-reward balance, where it has relatively higher average returns and moderate standard deviation.


#### Portfolio Risk and Return

```python
# Calculate covariance matrix
cov_matrix = returns.cov()

# Define portfolio weights
weights = np.array([1 / len(returns.columns) for _ in range(len(returns.columns))]) # Equal weights for simplicity

# Calculate portfolio returns and volatility
portfolio_return = np.dot(mean_returns, weights)
portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Print portfolio statistics
print("Portfolio Statistics:")
print("Expected Annual Return:", portfolio_return * 252)  # 252 trading days
print("Volatility (annualized):", portfolio_volatility * np.sqrt(252))
print("Sharpe Ratio:", (portfolio_return - 0.03) / portfolio_volatility)  # Assuming risk-free rate of 3%

# Visualize cumulative returns
cumulative_returns = (1 + returns).cumprod()
cumulative_returns.plot(figsize=(10, 6))
plt.title('Cumulative Portfolio Returns')
plt.xlabel('date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid(True)
plt.show()
```

#### Key takeaways

- This portfolio on average, can expect to return about 13.05% per year.
- Over the year this portfolio is expected to fluctuate by 10.89%.
- It has a negative Sharpe ration of -4.30, indication the the portfolio returns are not compensation for the risk taken.

In conclusion the Sharpe ration indicate the this portfolio is performing poorly.Relatively to its risk the returns are low. the strategy is not effective in compensating for the risk taken.


#### Portfolio Optimization


```python
# Calculate daily returns
returns = low_data.pivot(columns='name', values='close').pct_change()

# Define risk-free rate
risk_free_rate = 0.03

# Define objective function (negative Sharpe ratio)
def negative_sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate):
    portfolio_return = np.dot(mean_returns, weights)
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility
    return -sharpe_ratio

# Define constraints (weights sum to 1)
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Define bounds (weights between 0 and 1)
bounds = tuple((0, 1) for _ in range(len(returns.columns)))

# Initialize equal weights as initial guess
initial_guess = np.array([1 / len(returns.columns) for _ in range(len(returns.columns))])

# Perform optimization
optimal_weights = minimize(negative_sharpe_ratio, initial_guess,
                           args=(returns.mean(), returns.cov(), risk_free_rate),
                           method='SLSQP', bounds=bounds, constraints=constraints)

# Print optimized portfolio weights
print("Optimized Portfolio Weights:")
for i, name in enumerate(returns.columns):
    print(name, ":", round(optimal_weights.x[i], 4))

# Calculate optimized portfolio statistics
optimal_portfolio_return = np.dot(returns.mean(), optimal_weights.x)
optimal_portfolio_volatility = np.sqrt(np.dot(optimal_weights.x.T, np.dot(returns.cov(), optimal_weights.x)))
optimal_sharpe_ratio = (optimal_portfolio_return - risk_free_rate) / optimal_portfolio_volatility

# Print optimized portfolio statistics
print("\nOptimized Portfolio Statistics:")
print("Expected Annual Return:", optimal_portfolio_return * 252)  # 252 trading days
print("Volatility (annualized):", optimal_portfolio_volatility * np.sqrt(252))
print("Sharpe Ratio:", optimal_sharpe_ratio)

# Visualize cumulative returns
cumulative_returns = (1 + returns.dot(optimal_weights.x)).cumprod()
cumulative_returns.plot(figsize=(10, 6))
plt.title('Optimized Portfolio Cumulative Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.grid(True)
plt.show()


```
